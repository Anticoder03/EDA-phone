```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("mobiles1.csv")
```

```python
data['processor'].fillna("Unknown", inplace=True)
data['rating'].fillna(data['rating'].mean(),inplace=True)
data['reviews'].fillna('0',inplace = True)
data['warranty'].fillna("Unknown", inplace=True)
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
# convert battery column to numeric (mAh)
data['battery'] = data['battery'].str.split().str[0].astype(int)
data['display'] = data['display'].str.split().str[0].astype(float)
data['reviews'] = (
    data['reviews']
    .str.replace(',', '', regex=False)
    .str.split()
    .str[0]
    .astype(float)
)

data['RAM_GB'] = (
    data['memory']
    .str.split('|')
    .str[0]
    .str.split()
    .str[0]
    .astype(float)
)

data['ROM_GB'] = (
    data['memory']
    .str.split('|')
    .str[1]
    .str.split()
    .str[0]
    .astype(int)
)

data['primary_camera_MP'] = (
    data['camera']
    .str.split('|').str[0]          # take rear camera part
    .str.findall(r'\d+')
    .apply(lambda x: max(map(int, x)) if len(x) > 0 else None)
)

data['front_camera_MP'] = (
    data['camera']
    .str.split('|').str[1]
    .str.findall(r'\d+')
    .apply(lambda x: max(map(int, x)) if isinstance(x, list) and len(x) > 0 else None)
)

```

```python
data['brand'] = data['name'].str.split().str[0].astype('str')
```

```python
data
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()

print(data['price'].mean() > data['price'].median())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['rating'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

print(f'Is the col right skewed {(data['rating'].mean() > data['rating'].median())}')
```

```python
cols = []
for i in data.columns:
     if data[i].dtype != 'object':
        cols.append(i)
cols
```

```python

numeric_cols = data.select_dtypes(exclude='object').columns

for col in numeric_cols:
    # plt.figure(figsize=(7, 4))
    sns.histplot(data[col], bins=30, kde=True)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

    mean_val = data[col].mean()
    median_val = data[col].median()

    if mean_val > median_val:
        print(f'{col}: Right-skewed')
    elif mean_val < median_val:
        print(f'{col}: Left-skewed')
    else:
        print(f'{col}: Approximately symmetric')

    print('-' * 40)


```

```python
continuous_cols = ['battery', 'display', 'rating']

for col in continuous_cols:
    plt.figure(figsize=(7, 4))
    sns.scatterplot(x=col, y='price', data=data)
    plt.title(f'Price vs {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    sns.regplot(x=col, y='price', data=data, scatter=True, line_kws={'color':'red'})

    plt.show()
```

```python
discrete_cols = ['RAM_GB', 'ROM_GB', 'primary_camera_MP']

for col in discrete_cols:
    plt.figure(figsize=(7, 4))
    sns.boxplot(x=col, y='price', data=data)
    plt.title(f'Price vs {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.show()
```

```python
top_brands = data['brand'].value_counts().head(10).index
brand_data = data[data['brand'].isin(top_brands)]
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='brand', y='price', data=brand_data)
plt.title('Price Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
corr_matrix = data.select_dtypes(exclude='object').corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Heatmap")
plt.show()
```

## âœ… Final EDA Summary

- **ROM** shows the strongest positive correlation with price.
- **RAM** contributes to price but with high variability at higher values.
- **Battery, display size, and camera resolution** show weak to moderate influence.
- **User ratings** are largely independent of price.
- **Brand positioning** explains many pricing outliers.
